```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,100)
```

```python
y=x*2
z=x**2
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
ax.plot(x,y,color='b')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax1.plot(x,y,color='r')
ax2.plot(x,y,color='r')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax1.plot(x,z,'b')
ax2.plot(x,y,'b')
ax2.set_xlim(20.0,22.0)
ax2.set_ylim(30,50)
ax2.set_title('Zoom')
```

```python
plt.subplot(1,2,1)
plt.subplot(1,2,2)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b',linestyle='dashed')
plt.subplot(1,2,2)
plt.plot(x,z,color='r',linewidth=4)
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,2))
plt.subplot(1,2,1)
plt.plot(x,y,'b' ,linestyle='solid',linewidth=4)
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(1,2,2)
plt.plot(x,z,'r--')
plt.xlabel('x')
plt.ylabel('z')
```
